```python
import higlass as hg

# start fuse process in background
hg.fuse.start("/content/tmp")

# remote bigwig
url = "http://hgdownload.cse.ucsc.edu/goldenpath/hg19/encodeDCC/wgEncodeSydhTfbs/wgEncodeSydhTfbsGm12878InputStdSig.bigWig"

# get mapping to local filesystem
path = hg.fuse.path(url)
ts = hg.bigwig(path)

hg.view(ts.track())
```

```python
hg.fuse.stop()  # manually stop, but will also stop when jupyter notebook is closed...
```
